```python
import matplotlib.pyplot as plt
import numpy as np
import ipywidgets as widgets
from IPython.display import display

# Function to update the radar chart based on the proportions
def update_radar_chart(TotalNumberUser, TotalProfit, TotalCost, TotalSales, Total5Star):
    # Convert input to proportions
    total_proportion = TotalNumberUser + TotalProfit + TotalCost + TotalSales + Total5Star
    TotalNumberUser = TotalNumberUser / total_proportion * 100
    TotalProfit = TotalProfit / total_proportion * 100
    TotalCost = TotalCost / total_proportion * 100
    TotalSales = TotalSales / total_proportion * 100
    Total5Star = Total5Star / total_proportion * 100

    # Calculate the values for each product based on the adjusted proportions
    values_A = [TotalNumberUser * 0.8 / 100, TotalProfit * 0.7 / 100, TotalCost * 0.6 / 100, TotalSales * 0.9 / 100, Total5Star * 0.8 / 100]
    values_B = [TotalNumberUser * 0.6 / 100, TotalProfit * 0.9 / 100, TotalCost * 0.5 / 100, TotalSales * 0.7 / 100, Total5Star * 0.6 / 100]
    values_C = [TotalNumberUser * 0.7 / 100, TotalProfit * 0.8 / 100, TotalCost * 0.7 / 100, TotalSales * 0.8 / 100, Total5Star * 0.9 / 100]

    # Close the loop for each set of values
    values_A += [values_A[0]]
    values_B += [values_B[0]]
    values_C += [values_C[0]]

    # Create a figure and axis for the radar chart
    fig, ax = plt.subplots(figsize=(8, 8), subplot_kw={'polar': True})

    # Plot the radar chart for each product
    ax.plot(angles, values_A, 'o-', linewidth=2, label=product_labels[0])
    ax.fill(angles, values_A, alpha=0.25)

    ax.plot(angles, values_B, 'o-', linewidth=2, label=product_labels[1])
    ax.fill(angles, values_B, alpha=0.25)

    ax.plot(angles, values_C, 'o-', linewidth=2, label=product_labels[2])
    ax.fill(angles, values_C, alpha=0.25)

    # Set the labels for each axis
    ax.set_xticks(angles[:-1])  # Exclude the last angle to avoid duplication
    ax.set_xticklabels(features)

    # Add a legend
    ax.legend(loc='upper right', bbox_to_anchor=(1.3, 1.0))

    # Show the radar chart
    plt.title('Product Advantages Radar Chart')
    plt.show()

# Create input boxes for each feature to adjust the proportion
TotalNumberUser_input = widgets.FloatText(value=60, min=0, max=100, description='TotalNumberUser (%):')
TotalProfit_input = widgets.FloatText(value=10, min=0, max=100, description='TotalProfit (%):')
TotalCost_input = widgets.FloatText(value=10, min=0, max=100, description='TotalCost (%):')
TotalSales_input = widgets.FloatText(value=10, min=0, max=100, description='TotalSales (%):')
Total5Star_input = widgets.FloatText(value=10, min=0, max=100, description='Total5Star (%):')

# Function to update other inputs based on the current input
def update_others(change):
    total_percentage = TotalNumberUser_input.value + TotalProfit_input.value + TotalCost_input.value + TotalSales_input.value + Total5Star_input.value
    remaining_percentage = 100 - total_percentage
    if remaining_percentage > 0:
        remaining_input = change.owner
        other_inputs = [TotalNumberUser_input, TotalProfit_input, TotalCost_input, TotalSales_input, Total5Star_input]
        other_inputs.remove(remaining_input)
        for inp in other_inputs:
            inp.max = remaining_percentage

# Observe changes in each input and update others
TotalNumberUser_input.observe(update_others, 'value')
TotalProfit_input.observe(update_others, 'value')
TotalCost_input.observe(update_others, 'value')
TotalSales_input.observe(update_others, 'value')
Total5Star_input.observe(update_others, 'value')

# Call the update function when inputs change
widgets.interactive(update_radar_chart,
                    TotalNumberUser=TotalNumberUser_input,
                    TotalProfit=TotalProfit_input,
                    TotalCost=TotalCost_input,
                    TotalSales=TotalSales_input,
                    Total5Star=Total5Star_input)
```
